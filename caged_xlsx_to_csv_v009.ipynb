```python
import pandas as pd
```

```python
# def caged_xlsx_to_csv():
# caged_xlsx_to_csv()

file_in = '3-tabelas.xlsx'
file_out = 'caged.csv'

skiprows = 4  # [0, 0, 0, 0, 0, 0, 0, 4, 0, 0, 0, 0, 0, 0]

# Base com todas as planilhas
dfb = pd.read_excel(file_in, sheet_name=None, skiprows=skiprows)
print(dfb.keys())
```

```python
df = dfb['Tabela 5.1']
df
```

```python
MONTHBR = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
MONTHNUMBER =  ['01',      '02',        '03',    '04',    '05',   '06',    '07',    '08',     '09',       '10',      '11',       '12']
mbr2mnumber = dict(zip(MONTHBR, MONTHNUMBER))
df2 = df.copy()
df2 = df2[ (df2['Mês'].str.slice(-4) >= '1900') & (df2['Mês'].str.slice(-4) <= '2099') ]
df2 = df2.dropna(how = 'all')

df2['Mês'] = df2['Mês'].replace(mbr2mnumber, regex=True)
df2['Mês'] = df2['Mês'].str.replace('/','-')
df2['Mês'] = pd.to_datetime(df2['Mês'], format = '%Y-%m-%d', infer_datetime_format = True)

df2['Saldos'] = df2['Saldos'].round().astype(int)

df2[['Mês', 'Saldos']].to_csv(file_out, index=False)

df2.rename(columns = {'Mês':'', 'Saldos':'total'}, inplace = True)
df2[['', 'total']].to_csv(file_out, index=False)
df2[['','total']]
```
